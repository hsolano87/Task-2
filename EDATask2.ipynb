```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credit = pd.read_csv('default of credit card clients.csv', header =1)
```

```python
header = credit.dtypes.index
print(header)
```

```python
# Caracter√≠sticas de las variables
credit.info()
```

```python
#Cantidad de clientes que pagaran=0 o que no pagaran=1
plt.hist(credit['default payment next month'])
```

```python
# Cantidad de hombre y mujeres
plt.hist(credit['SEX'])
```

```python
## 77,9%  pagara y 22,1% no pagara
## 39,6 % de hombre y 60,4% de hombres
#Quienes pagan a tiempo, hombre o mujeres, cantidad y pago
sns.catplot('SEX', data=credit, kind='count', aspect=1.5)
```

```python
# Seaborn para agrupar por sexo quien paga o no el proximo mes
g = sns.catplot('default payment next month', data=credit, hue='SEX', kind='count', aspect=1.75)
g.set_xlabels('default payment next month')
```

```python
# Cantidad de credito por nivel de educacion
sns.catplot('EDUCATION', data=credit, kind='count', aspect=1.5)
```
